## Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('ggplot')
```

```python
print(plt.style.available)
```

### Load Data

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
df.shape
```

```python
df.isna().any()
```

```python
df.describe()
```

### Data Visualization

#### Histogram - Distribution

```python
plt.hist(df['sepal_length'])
```

```python
df['sepal_length'].hist()
plt.xlabel('Sepal Length')
plt.ylabel('Density')
```

```python
df['petal_length'].hist()
plt.xlabel('Sepal Length')
plt.ylabel('Density')
```

```python
df['species'].hist()
plt.xlabel('Species')
plt.ylabel('Density')
```

```python
sns.countplot(x=df['species'])
```

#### Scatter Plot - Distribution & Relationship

```python
plt.scatter(x = df['sepal_length'], y = df['sepal_width'])

plt.tight_layout()
plt.show()
```

```python
df.plot(kind='scatter', x='sepal_length', y='sepal_width')
```

```python
df.plot(kind='scatter', x='petal_length', y='petal_width')
```

```python
sns.scatterplot(data=df, x=df['petal_length'], y=df['petal_width'], hue=(df['species']), palette='Set2')
```

#### Distplot - Distribution

```python
sns.distplot(df['sepal_length'])
```

```python
sns.kdeplot(x=df['sepal_length'], hue=df['species'], shade=True)
```

#### Boxplot - Distribution & Comparison

```python
sns.boxplot(data=df)
plt.xlabel('Species')
plt.ylabel('Length in cm')
plt.show()
```

```python
sns.boxplot(data=df, x=df['species'], y=df['petal_length'])
plt.xlabel('Species')
plt.ylabel('Length in cm')
plt.show()
```

#### Pairplot - Relationship, Distribution, Comparison

```python
sns.pairplot(data=df, hue='species')
plt.show()
```

#### Heatmap - Correlation/Relationship

```python
df.corr()
```

```python
sns.heatmap(data=df.corr(), annot=True)
plt.show()
```
